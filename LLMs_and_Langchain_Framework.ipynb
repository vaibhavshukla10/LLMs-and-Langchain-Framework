1.  What are Large Language Models (LLMs) and how do they function?

      - Large Language Models (LLMs) are advanced Artificial Intelligence (AI) models trained on massive amounts of text data. They are designed to understand, generate, summarize, translate, and answer questions in human language.

     **How LLMs Function**

       a} Training on Large Datasets -

       LLMs are trained using billions of words from books, websites, articles, and other text sources.
       During training, they learn grammar, vocabulary, facts, reasoning patterns, and language structure.

       b} Tokenization -  Input text is divided into small units called tokens (words or parts of words).
            Example:
          "The sky is blue" → ["The", "sky", "is", "blue"]

       c} Transformer Architecture - LLMs use the Transformer neural network architecture.
          A mechanism called self-attention helps the model understand relationships between words in a sentence

       d} Prediction - The model predicts the most likely next token based on the previous tokens.
        By repeating this process, it generates complete sentences and paragraphs.

      e} Fine-Tuning (Optional) - After initial training, the model can be fine-tuned on specific datasets for tasks such as chatbots, coding, translation, or medical applications.

      **Applications**

       -  Chatbots and virtual assistants
       - Text generation
       - Language translation
       - Text summarization
       - Code generation
       - Question answering

# Compare and Contrast: LangChain vs LlamaIndex

## Introduction

LangChain and LlamaIndex are popular frameworks used for building applications with Large Language Models (LLMs). Although they are often used together, they solve different problems in LLM-based systems.

---

# 1. LangChain

**LangChain** is a framework used to build complete LLM applications by connecting multiple components together.

### Main Focus:

* Building workflows using LLMs
* Creating AI agents
* Connecting tools, APIs, and databases

### What it solves:

LangChain solves the problem of:

> “How can we build intelligent applications that can think step-by-step, use tools, and perform complex tasks?”

### Key Features:

* Chains (step-by-step processing)
* Agents (decision-making AI systems)
* Tool integration (Google search, APIs, calculators, etc.)
* Multi-step reasoning pipelines

### Example:

A chatbot that:

1. Understands user question
2. Searches web or database
3. Uses tools (like calculator or API)
4. Gives final answer

---

# 2. LlamaIndex

**LlamaIndex** is a framework designed to connect LLMs with external/private data sources.

### Main Focus:

* Data indexing and retrieval
* Building knowledge-based systems
* Supporting RAG (Retrieval-Augmented Generation)

### What it solves:

LlamaIndex solves the problem:

> “How can we make LLMs answer questions using our own documents and data?”

### Key Features:

* Document loading (PDFs, text files, websites)
* Indexing and structuring data
* Fast retrieval of relevant information
* RAG pipeline support

### Example:

A chatbot that answers questions from:

* Company documents
* PDFs
* Research papers
* Internal knowledge base

---

# 3. Comparison Table

| Feature             | LangChain               | LlamaIndex                |
| ------------------- | ----------------------- | ------------------------- |
| Main Purpose        | Build LLM applications  | Connect LLMs to data      |
| Focus Area          | Workflows & agents      | Data retrieval & indexing |
| Tool integration    | Strong                  | Limited                   |
| RAG support         | Possible                | Core strength             |
| Document handling   | Basic                   | Advanced                  |
| Complexity handling | High (multi-step logic) | Medium (data-focused)     |

---

# 4. Key Differences

### LangChain:

* Focuses on **logic and workflow**
* Used to build **AI agents and complex applications**
* Works like a **control system for LLM apps**

### LlamaIndex:

* Focuses on **data and knowledge**
* Used to build **question-answering systems**
* Works like a **bridge between data and LLMs**

---

# 5. How They Work Together

In many real-world applications, both are used together:

* **LlamaIndex** → Fetches relevant data from documents
* **LangChain** → Uses that data to perform reasoning and generate final output

---

# Conclusion

LangChain is best for building **complex LLM-based applications with workflows and agents**, while LlamaIndex is best for **connecting LLMs to custom data for efficient retrieval and question answering**. Both tools complement each other in modern AI systems.

7.  Integrate Langchain with a third-party API (e.g., weather, news) and show
how responses can be generated via LLMs.
(Include your Python code and output in the code box below.)

       - **🌦️ LangChain + Weather API Integration with LLM**


```python
# Install libraries
!pip install langchain openai requests langchain-openai

import os
import requests
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# API Keys
os.environ["OPENAI_API_KEY"] = "your_openai_api_key"
WEATHER_API_KEY = "your_openweather_api_key"

# Weather Function
def get_weather(city):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={WEATHER_API_KEY}&units=metric"
    data = requests.get(url).json()
    return {
        "city": city,
        "temperature": data["main"]["temp"],
        "humidity": data["main"]["humidity"],
        "condition": data["weather"][0]["description"]
    }

# LLM Setup
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)

# Prompt Template
prompt = ChatPromptTemplate.from_template("""
Weather Data:
City: {city}
Temperature: {temperature}°C
Humidity: {humidity}%
Condition: {condition}

Explain this weather in simple words.
""")

# Chain
chain = prompt | llm | StrOutputParser()

# Input + Output
city = input("Enter city: ")
weather = get_weather(city)
result = chain.invoke(weather)

print("\nAI Weather Report:\n")
print(result)
```

8. Create a LamaIndex implementation that indexes a local text file and
retrieves answers from it.
(Include your Python code and output in the code box below.)
         
      **📚 LlamaIndex: Index Local Text File & Ask Questions**

```python
# Install required libraries
!pip install llama-index openai

import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.openai import OpenAI

# Set OpenAI API Key
os.environ["OPENAI_API_KEY"] = "your_api_key_here"

# -----------------------------
# Step 1: Create a sample text file
# -----------------------------
text_content = """
Artificial Intelligence (AI) is a field of computer science focused on creating systems
that can perform tasks requiring human intelligence. Machine Learning is a subset of AI
that enables systems to learn from data. Deep Learning is a subset of Machine Learning
based on neural networks with many layers.
"""

with open("data.txt", "w") as f:
    f.write(text_content)

# -----------------------------
# Step 2: Load document
# -----------------------------
documents = SimpleDirectoryReader(input_files=["data.txt"]).load_data()

# -----------------------------
# Step 3: Create Index
# -----------------------------
index = VectorStoreIndex.from_documents(documents)

# -----------------------------
# Step 4: Create Query Engine
# -----------------------------
query_engine = index.as_query_engine()

# -----------------------------
# Step 5: Ask Questions
# -----------------------------
question = input("Ask a question from the document: ")

response = query_engine.query(question)

print("\n=== Answer from LlamaIndex ===\n")
print(response)
```

9. Demonstrate combining Langchain with LamaIndex to create a simple
document-based Q&A chatbot.
(Include your Python code and output in the code box below.)

      **🤖 LangChain + LlamaIndex Document Q&A Chatbot**

```python
# Install required libraries
!pip install llama-index langchain openai

import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Set API Key
os.environ["OPENAI_API_KEY"] = "your_api_key_here"

# -----------------------------
# Step 1: Create a local document
# -----------------------------
text = """
Python is a high-level programming language.
It is widely used in AI, machine learning, and web development.
LangChain helps build LLM applications using chains and agents.
LlamaIndex helps connect LLMs with external documents for retrieval-based QA systems.
"""

with open("doc.txt", "w") as f:
    f.write(text)

# -----------------------------
# Step 2: Load document using LlamaIndex
# -----------------------------
documents = SimpleDirectoryReader(input_files=["doc.txt"]).load_data()
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

# -----------------------------
# Step 3: Ask question using LlamaIndex (retrieval)
# -----------------------------
question = input("Ask your question: ")
retrieved_answer = query_engine.query(question)

# -----------------------------
# Step 4: Use LangChain to format final response
# -----------------------------
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.5)

prompt = ChatPromptTemplate.from_template("""
You are a helpful assistant.

Context from document:
{context}

User Question:
{question}

Give a simple and clear final answer.
""")

chain = prompt | llm | StrOutputParser()

# -----------------------------
# Step 5: Combine LlamaIndex + LangChain
# -----------------------------
response = chain.invoke({
    "context": str(retrieved_answer),
    "question": question
})

print("\n=== Final Chatbot Answer ===\n")
print(response)
```

10. A legal firm wants to use AI to summarize large volumes of legal
documents and retrieve relevant information quickly. Propose a solution using
Langchain and LamaIndex, and explain how it would work in practice.
(Include your Python code and output in the code box below.)

    **⚖️ Legal AI System using LangChain + LlamaIndex (RAG Solution)**

```python
# Install required libraries
!pip install llama-index langchain openai

import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Set API Key
os.environ["OPENAI_API_KEY"] = "your_api_key_here"

# ------------------------------------
# Step 1: Create sample legal documents
# ------------------------------------
legal_text = """
Case 1: Contract Law
A contract is an agreement between two or more parties that is enforceable by law.
Breach of contract occurs when one party fails to fulfill obligations.

Case 2: Criminal Law
Criminal law deals with crimes against society.
Punishments include fines, imprisonment, or both.

Case 3: Intellectual Property
Intellectual property law protects creations like inventions, trademarks, and copyrights.
"""

with open("legal_docs.txt", "w") as f:
    f.write(legal_text)

# ------------------------------------
# Step 2: Load documents using LlamaIndex
# ------------------------------------
documents = SimpleDirectoryReader(input_files=["legal_docs.txt"]).load_data()

index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

# ------------------------------------
# Step 3: Retrieve relevant context
# ------------------------------------
question = input("Enter legal question: ")
retrieved_info = query_engine.query(question)

# ------------------------------------
# Step 4: Use LangChain for final summarization
# ------------------------------------
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.3)

prompt = ChatPromptTemplate.from_template("""
You are an AI assistant for a legal firm.

Use the following retrieved legal context:
{context}

Question:
{question}

Provide a clear, short, and professional legal summary answer.
""")

chain = prompt | llm | StrOutputParser()

# ------------------------------------
# Step 5: Generate final response
# ------------------------------------
final_answer = chain.invoke({
    "context": str(retrieved_info),
    "question": question
})

print("\n=== Legal AI Response ===\n")
print(final_answer)
```
